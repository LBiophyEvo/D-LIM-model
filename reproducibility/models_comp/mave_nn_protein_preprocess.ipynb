```python
# Load data from xlsx
df_1 = pd.read_excel("../../data/protein_inter/GSE102901_TableS1.xlsx", sheet_name=None)
# df_2 = pd.read_excel("./data/protein_inter/GSE102901_TableS4.xlsx", sheet_name=None)

# Select data from sheet 2 
tmp = df_1["Sheet2"][["id1", "id2", "pos1", "mut1", "pos2", "mut2", "d_mean", "epi"]]

# Only keep the fitness and epistasis data  
# new_table = tmp[["id1", "id2", "epi"]]
# new_table.to_csv("../data/protein_inter/tables1_epi.csv", index=False, header=False)
# new_table = tmp[["id1", "id2", "d_mean"]]
# new_table.to_csv("../data/protein_inter/tables1.csv", index=False, header=False)
```

```python
df_1['Sheet2']
```

```python
# preprocessing for protein 
len_gen1 = tmp['pos1'].max() 

tmp['mut1_sub'] = df_1['Sheet2']['wt1'].astype(str) + df_1['Sheet2']['pos1'].astype(str) + df_1['Sheet2']['mut1'].astype(str)
tmp['pos2_ext'] = df_1['Sheet2']['pos2'] + len_gen1
tmp['mut2_sub'] = df_1['Sheet2']['wt2'].astype(str) + tmp['pos2_ext'].astype(str) + df_1['Sheet2']['mut2'].astype(str)
tmp['substition'] = tmp['mut1_sub'] + ':' + tmp['mut2_sub']
tmp
```

```python
fos_widetype = [[] for el in range(len_gen1)]
for mut, pos in zip(df_1['Sheet2']['wt1'],df_1['Sheet2']['pos1']): 
    fos_widetype[(pos-1)] = mut
fos_widetype

jun_widetype = [[] for el in range(len_gen1)]
for mut, pos in zip(df_1['Sheet2']['wt2'],df_1['Sheet2']['pos2']): 
    jun_widetype[(pos-1)] = mut
jun_widetype

widetype = fos_widetype + jun_widetype
widetype = ''.join(widetype)
print(widetype)
```

```python
def func_seq(pos1, mut1, pos2, mut2):
    seq = list(widetype)
    seq[pos1-1] = mut1
    seq[pos2-1 + len_gen1] = mut2
    return ''.join(seq)
tmp['x'] = tmp.apply(lambda row: func_seq(row['pos1'], row['mut1'], row['pos2'], row['mut2']), axis = 1)
```

```python
df = tmp[['x', 'epi']]
df.columns = ['x', 'y']
df = df.dropna()
df.to_csv('../../data/protein_inter/protein_mevv_epi.csv', index =None)
df
```

```python
df = tmp[['x', 'd_mean']]
df.columns = ['x', 'y']
df = df.dropna()
df.to_csv('../../data/protein_inter/protein_mevv_fitness.csv', index =None)
df
```
