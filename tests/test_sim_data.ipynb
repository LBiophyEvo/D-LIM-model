# Test the landscape generated by sim_data.py 

```python
import sys 
sys.path.append('../')
from src_simulate_data.sim_data import Simulated 
import matplotlib.pyplot as plt 
```

```python
fig, axes = plt.subplots(3, 3, figsize=(9, 9))

cor_types= ['exp', 'tgaus', 'cascade', 'bio', 'add', 'quad', 'comp', 'saddle', 'hat']
for id, cor in enumerate(cor_types):
    # simulate the landscape
    data_simulated = Simulated(nb_var=30, cor=cor)

    # Visualize the landscape 
    id_x, id_y = id//3, id%3
    data_simulated.plot(axes[id_x, id_y])
    axes[id_x, id_y].set_title(f'{cor}')

plt.tight_layout()
plt.show()
```
